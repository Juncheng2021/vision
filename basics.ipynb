```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, io
import os
```

```python
# Reading and displaying an image with plt
dirname= os.path.abspath('')
mountains = io.imread(os.path.join(dirname, 'images/ali-kazal.jpg'))
plt.imshow(mountains)
```

```python
print("Shape: ", mountains.shape)
print("Size: ", mountains.size)
print("Array:\n", mountains)
```

```python
def multiply_colour(im, index, scalar):
    f, ax = plt.subplots(1,2)
    ax[0].imshow(im)

    # Image is a 3D numpy array. It holds arrays (representing rows) of arrays (representing pixels) with 3 values each for RGB
    mult = np.copy(im)
    mult[:,:,index] *= scalar

    ax[1].imshow(mult)

im = data.rocket()
multiply_colour(im, 0, 2)
multiply_colour(im, 1, 2)
multiply_colour(im, 2, 2)
```

```python
# Numpy allows us to apply comparison operators to elements of arrays and then target those elements that come out as true
hello = np.array([1,2,3,4,5])
ufunc = hello < 3
hello[ufunc] = 0
print(hello)

# This can be used for thresholding
im = data.camera()
print("Shape:", im.shape)
print("Pixels:", im)

# Camera image black and white, just monochrome values 0->255
f, ax = plt.subplots(1,2)
ax[0].imshow(im, cmap='gray')
mask = im < 100
not_mask = im > 100
im[mask] = 255
im[not_mask] = 0
ax[1].imshow(im, cmap='gray')
```
